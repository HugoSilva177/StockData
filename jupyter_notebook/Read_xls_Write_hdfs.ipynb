```python
papel = 'VALE3'
file_loc = './Balancos_xls/%s_31.12.2020.xls' % papel
```

## Leitura da sheet Balanço Patrimonial

```python
balanco_patrimonial_df = pd.read_excel(file_loc, sheet_name='Bal. Patrim.', skiprows=[0], na_values=['NA'])
balanco_patrimonial_df.head()
```

```python
balanco_patrimonial_df = balanco_patrimonial_df.rename(columns={list(balanco_patrimonial_df)[0]: "Tipo"})
balanco_patrimonial_df.head()
```

## Leitura da sheet Demonstrativo de Resultados

```python
demonstrativo_resultado_df = pd.read_excel(file_loc, sheet_name='Dem. Result.', skiprows=[0], na_values=['NA'])
demonstrativo_resultado_df = demonstrativo_resultado_df.rename(columns={list(demonstrativo_resultado_df)[0]:"Tipo"})
```

```python
demonstrativo_resultado_df.head()
```

# Acesso ao HDFS server (Write and Read)
## Criando um Spark Session

```python
from pyspark.sql import SparkSession
sparkSession = SparkSession.builder.appName("write-read-balancos-dre-xls-hdfs").getOrCreate()
```

## Write Spark DataFrame do Balanço Patrimonial

```python
balanco_patrimonial_label = list(balanco_patrimonial_df.columns)
```

```python
balanco_patrimonial_dados = list(map(lambda lista: tuple(lista), balanco_patrimonial_df.values))
```

```python
df = sparkSession.createDataFrame(balanco_patrimonial_dados, balanco_patrimonial_label)
df
```

```python
df.write.save('hdfs://172.17.177.40:9000/user/hadoopuser/fundamentus/balanco/vale3', format='parquet', mode='append')
```

```python
df_load = sparkSession.read.format('parquet').load('hdfs://172.17.177.40:9000/user/hadoopuser/fundamentus/balanco/vale3')
```

```python
df_load.show()
```

```python
df_load.toPandas().head()
```
